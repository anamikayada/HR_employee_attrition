# 2)Monthly Income Trend with respect to Attrition


```python
plt.figure(figsize=(20,10),facecolor='white')
ax=sns.histplot(data=df,x="MonthlyIncome",hue="Attrition",palette='rocket')

for container in ax.containers:
        ax.bar_label(container, fmt="%d", label_type="edge")
plt.show()
```

By the plot we can conclude that there are 115 empolyee who has monthly income between 2400 to 2600 doller are more willing to leave<br>

Yes, we can conclude that attrition is indirectly proportional to monthly income. This means that as monthly income increases, attrition"yes" decreases. This is because employees with higher incomes are more likely to be satisfied with their jobs and less likely to leave for other opportunities.

The fact that employees having a monthly income between 15,000 and 18,000 doller have zero attrition"yes" supports this conclusion. It suggests that this income range is a threshold where employees are most likely to be satisfied with their jobs and stay with the company.
